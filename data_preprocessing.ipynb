```python
import pandas as pd
import os
import shutil
```

```python
train_df = pd.read_csv('train_2.csv')
test_df = pd.read_csv('test_2.csv')
```

```python
# Assume train_df and test_df already exist and have an 'img_path' column
# Sample 1/3 of the training data
# train_df_sampled = train_df.sample(frac=1/3, random_state=42).reset_index(drop=True)
# test_df_sampled = test_df.sample(frac=1/3, random_state=42).reset_index(drop=True)

# Create new directories for sampled images
os.makedirs("sampled_images2/train", exist_ok=True)
os.makedirs("sampled_images2/test", exist_ok=True)

# Function to copy images to the new folder and update img_path
def copy_images_and_update_paths(df, new_folder):
    new_img_paths = []
    for i, row in df.iterrows():
        original_path = 'figures/' + row['Figure_path']
        img_name = os.path.basename(original_path)
        new_path = os.path.join(new_folder, img_name)
        
        # Copy image if it does not already exist in the new folder
        if not os.path.exists(new_path):
            shutil.copy(original_path, new_path)
        
        new_img_paths.append(new_path)
    
    df['img_path'] = new_img_paths
    return df

# Copy the sampled training images
train_df_sampled = copy_images_and_update_paths(train_df, "sampled_images2/train")

# Copy the sampled test images
test_df_sampled = copy_images_and_update_paths(test_df, "sampled_images2/test")

# Optionally, you could save these DataFrames to new CSVs for future reference:
train_df.to_csv("train_df.csv", index=False)
test_df.to_csv("test_df.csv", index=False)
```
